# EDA

```python
import pandas as pd
import numpy as np
#from sklearn.cluster import KMeans
```

```python
#data = pd.read_csv('final_census_data.csv', index_col = 0)
data = pd.read_csv('final_census_cleaned.csv')
```

```python
data.head()
```

```python
cols = data.columns.to_list()
objecttrue = ['GEO_ID','COUNTY','STATE']
objectcol = []
for i in cols:
    typ = data[i].dtypes
    if typ == 'O':
        objectcol.append(i)
```

```python
objectwrong = [i for i in objectcol if i not in objecttrue]
data[objectwrong].info()
```

```python
data[objectwrong] = data[objectwrong].astype(str)
'.' in data['WORKERS PER CAR, TRUCK, OR VAN'][1]
```

```python
for i in ['WORKERS PER CAR, TRUCK, OR VAN','MEAN TRAVEL TIME TO WORK (MINUTES)']:
    data[i] = data[i].apply(lambda x: float(x) if str(x).isalpha() == False else np.nan)
    missing = data[i].isna().sum()
    print(i+' | missing: '+ str(missing))
```

```python
objectwrong.remove('WORKERS PER CAR, TRUCK, OR VAN')
objectwrong.remove('MEAN TRAVEL TIME TO WORK (MINUTES)')
```

```python
objectwrong
```

```python
for num,i in enumerate(objectwrong):
    data[i] = data[i].apply(lambda x: np.nan if '-' in str(x) else float(x))
    missing = data[i].isna().sum()
    print(i+' | missing: '+ str(missing))
```

```python
objectwrong.append('WORKERS PER CAR, TRUCK, OR VAN')
objectwrong.append('MEAN TRAVEL TIME TO WORK (MINUTES)')
```

```python
for num,i in enumerate(data.columns.to_list()):
    missing = data[i].isna().sum()
    if missing > 0:
        print(i+' | missing: '+ str(missing))
```

# Impute missing values (KNN)

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=1)
```

```python
data = data.drop('Unnamed: 0',axis=1)
cols = data.columns.to_list()
numcol =cols[3:]
numcol
```

```python
data[numcol] = imputer.fit_transform(data[numcol])
```
