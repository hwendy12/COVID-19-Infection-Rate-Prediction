```python
# process GEO_Morbility data 
import pandas as pd 
import numpy as np
import os

from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
```

```python
# geo_morbility data 
gm_path = path + "/gm/gm_data.csv"
gm_data = pd.read_csv(gm_path)
```

```python
feature_list = [name for name in gm_data.iloc[0,:] if "Moved" not in name ]
feature_list = [name for name in feature_list if "Margin" not in name ]
feature_list = [name for name in feature_list if "CITIZENSHIP" not in name ]
feature_list = [name for name in feature_list if "RACE" not in name ]
feature_list = [name for name in feature_list if "SEX" not in name ]
feature_list = [name for name in feature_list if "MARITAL" not in name ]
feature_list = [name for name in feature_list if "Estimate!!Total!!Population" not in name ]
```

```python
final_feature_list = [name for name in feature_list if name not in['Estimate!!Total!!EDUCATIONAL ATTAINMENT!!Population 25 years and over',
                                                                   'Estimate!!Total!!INDIVIDUAL INCOME IN THE PAST 12 MONTHS (IN 2018 INFLATION-ADJUSTED DOLLARS)!!Population 15 years and over',
                                                                   'Estimate!!Total!!POVERTY STATUS IN THE PAST 12 MONTHS!!Population 1 year and over for whom poverty status is determined',
                                                                   'Estimate!!Total!!HOUSING TENURE!!Population 1 year and over in housing units',
                                                                   'Estimate!!Total!!PERCENT ALLOCATED!!Residence 1 year ago'
                                                                  ]]
```

```python
gm_data = pd.read_csv(gm_path, skiprows =[0])
```

```python
processed_gm_data = gm_data[final_feature_list]
```

```python
new_row_name = [name.split("!!")[-1] for name in final_feature_list]
processed_gm_data.columns = new_row_name
processed_gm_data.to_csv('processed_gm_data.csv',index = False)
```
